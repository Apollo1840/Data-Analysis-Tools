```python
import matplotlib.pyplot as plt
import numpy as np
import wave
import wfdb 
```

# .wav file

```python
with wave.open('datasets/sig100.wav','r') as f_sound:
    
    framerate = f_sound.getframerate()

    # Extract Raw Audio from Wav File
    signal = np.fromstring(f_sound.readframes(-1), 'Int16')

    # If Stereo
    # if f_sound.getnchannels() == 2:
    #    print ('Just mono files 11')

time = np.linspace(0, len(signal)/framerate, num=len(signal))
```

```python
plt.plot(time, signal)
plt.xlim([1, 15])
plt.title('Record 100 from MIT-BIH Arrhythmia Database .wav form')
plt.show()
```
